Is there a correlation between increased completion of Masters and PhD and increased completion of Bachelors Degrees over the last 47 years? Is there a distinction between race and degrees chosen over the past 4 years. 

## Proposition/Hypothesis: Due to increased specificity of jobs and variety available there will be a more even distribution between fields of study in all degrees. In addition due to there being more demanding skill speciality, many students will complete an increasing varied types of degrees over the years. The more students complete their Bachelors degree the higher the likelihood of them completing a Masters and PhD due to greater exposure to fields of interest for more students which would make them more likely to find something of great interest. Based on lack of resources and experience in family with college education, minorities will find it more difficult to take on a variety of degrees.

## ANALYSIS OF PART I: Trend in Bachelors, Masters, and PhD Degrees
    ### Upon observation of the data from the tables and the charts there can be plenty of support found for the hypothesis. In the initial table of correlation values between the years, there can an observation made that the values between the years have decreasing correlation showing that throughout the years students are graduating with varying Bachelors degrees even though the correlation is not that strong. For bachelors degrees every year there is a degrees. For masters degrees the distinction is greater with a decrease yearly. For PhD because it is more rigourous it appears that while there is a yearly regression it is not as strong as the other two. There can also be an observation made on the charts that there is a correlation between the maximum bachelors degrees resulting in subsequent increase in masters degrees in the upcoming years and maximum degrees in PhD in upcoming years. This supports the idea that the more students are exposed to field the more likely they are to pursue it and continue their education. It is also clear that there is a huge distinction between Bachelors degrees and decrease in number of Masters and subsequently a decrease in PhD's. 
    
    It was interesting to note that the most popular bachelors degree was in education between 1970-80 and from the 1980s til 2017 Business was the first choice bachelors degree. While both fields were the only two on top in Masters Degrees, Business degrees were on top only in 2010 and afterwards, much later than in Bachelors degrees. Another note of interest was that for PhD's neither business or education were preferred. Legal professions and studies dominated until 1996 and from 2000 on health professions and related programs were top PhD choices. 
    
    From the Graphs and charts we can see that the percentage top percentage of the 5 largest bachelors degerees accounted for about 65% whereas by 2016 only around 50% of degrees were of the top 5 fields. This means more students chose a greater variety of fields. For Masters degree there was a similar trend as the bachelors, supporting the hypothesis. For PhD it was not the same as the top 5 fields accounted for about 70% of all degrees and the number remained consistent throughout the years.
